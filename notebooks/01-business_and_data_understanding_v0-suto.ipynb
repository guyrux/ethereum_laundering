
# Data import
We retrieve all the required data for the analysis.

```python
try:
    df_raw_features = pd.read_parquet(INTERIM_FOLDER + 'elliptic_txs_features.pqt')
    df_raw_classes = pd.read_parquet(INTERIM_FOLDER + 'elliptic_txs_classes.pqt')
    df_raw_edgelist = pd.read_parquet(INTERIM_FOLDER + 'elliptic_txs_edgelist.pqt')
    print('PARQUET file loaded.')
except FileNotFoundError as e:
    df_raw_features = pd.read_csv(RAW_FOLDER + 'elliptic_txs_features.csv', header=None)
    df_raw_classes = pd.read_csv(RAW_FOLDER + 'elliptic_txs_classes.csv')
    df_raw_edgelist = pd.read_csv(RAW_FOLDER + 'elliptic_txs_edgelist.csv')
    
    lst_tx = ['tx_'+str(col).zfill(3) for col in df_raw_features.columns[2:95]]
    lst_agg = ['agg_'+str(col).zfill(3) for col in df_raw_features.columns[95:]]
    lst_columns = ['id', 'time_step', *lst_tx, *lst_agg]
    df_raw_features.columns = lst_columns
    
    df_raw_features.to_parquet(INTERIM_FOLDER +  'elliptic_txs_features.pqt', index=False)
    df_raw_classes.to_parquet(INTERIM_FOLDER + 'elliptic_txs_classes.pqt', index=False)
    df_raw_edgelist.to_parquet(INTERIM_FOLDER + 'elliptic_txs_edgelist.pqt', index=False)
    print('EXCEL file loaded and PARQUET created.')
```
